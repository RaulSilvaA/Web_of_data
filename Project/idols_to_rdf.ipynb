#Conectando los archivos a wikidata

Se desea realizar un mapeo de los nombres a su ID de wikidata, esto con el fin de poder acceder a información extra presente en esta, además de evitar problemas con grupos cuyos nombres poseen valores numéricos o caracteres especiales como "%" o "$".

Para esto se corre el siguiente script manualmente desde [wikidata](https://query.wikidata.org), donde el *input* no es más que el string que se genera a partir de *nameToWikidata*



```
SELECT * WHERE {   
  {?s rdfs:label|skos:altLabel ?label .   
  ?s wdt:P31/wdt:P279* wd:Q641066. }  
  UNION
  {?s rdfs:label|skos:altLabel ?label .   
  ?s wdt:P31/wdt:P279* wd:Q216337. } 
  VALUES ?label {  
    %input
  }
}
```
Este paso es necesario para las 2 entidades de bandas, pues para los *idols* se puedan linkear luego de manera correcta. Es por esto que se crea un único csv que contiene los IDs tanto de las girlbands como de las boybands.


##IDs código

````python
girl_band_name = girlGroups["Name"]
boy_band_name = boyGroups["Name"]

def nameToWikidata(names):
    names = names.map(' "{}"@en'.format)
    names_string = ""
    for name in names:
        names_string += name + "\n"
    return names_string

girl_names = nameToWikidata(girl_band_name)
boy_names = nameToWikidata(boy_band_name)
band_names = girl_names + boy_names
print(band_names.lower())

        
````

````python
def findTheQ(string):
    i = 0
    while string[-i] != "Q":
        i += 1
    return string[-i:]

#band_ids from wikidata

band_ids = pd.read_csv('https://raw.githubusercontent.com/RaulSilvaA/Web_of_data/main/Project/CSVdata/band_ids.csv')    
band_ids["wd_id"] = band_ids["wd_id"].map(findTheQ)

#band_ids2 from manual cleaning
band_ids2 = pd.read_csv('https://raw.githubusercontent.com/RaulSilvaA/Web_of_data/main/Project/CSVdata/band_ids2.csv')
band_ids2["wd_id"] = band_ids2["wd_id"].map(findTheQ)
````

##Idols

````python
def replacer(rdf):
    rdf = rdf.replace(' ', '')
    rdf = rdf.replace('&', 'And')
    rdf = rdf.replace('-', '')
    rdf = rdf.replace("'", '')
    rdf = rdf.replace('#', 'HT')
    rdf = rdf.replace(',', 'And')
    rdf = rdf.replace('+', 'Plus')
    rdf = rdf.replace('.', 'Dot')
    rdf = rdf.replace('(', '')
    rdf = rdf.replace(')', '')
    rdf = rdf.replace('/', 'Of')
    rdf = rdf.replace('!', '')
    rdf = rdf.replace("21","TwentyOne")
    return rdf
````

````python
#Obtiene el código de wikidata para el grupo de un idol y lo usa en la informacion del grupo para su rdf
def element_in_group_idols(element, dataset, dataset2, label):
    #element es el nombre del grupo del idol
    #dataset es el dataset de ids de las bandas
    #label es grupo u otro grupo dependiendo del campo en que se este usando
    clear_element = element.replace(" ", "")
    if element in dataset['label'].tolist():
        id_row = (dataset.loc[dataset['label'] == element])['wd_id'].values[0]
        text = '\t ex:'+ label + ' wd:' + id_row + '; \n'
    elif clear_element in dataset2['label'].tolist():
        id_row = (dataset2.loc[dataset2['label'] == clear_element])['wd_id'].values[0]
        text = '\t ex:'+ label + ' wd:' + id_row + '; \n'
    else:
        text = '\t ex:'+ label + ' ex:' + replacer(clear_element) + '; \n'
    return text


#agrega un punto al final del archivo rdf
def point_end(s):
    if s[-3] == ";":
        inf = s[:-3]
        sup = s[-2:]
        s = inf + "." + sup 
    return s

def idolToRDF(dataset, bands_ids_dataset, band_ids_dataset2):
    rdf = ''
    index = 0 #solver for index problem
    for name in dataset['Full Name']: 
        #Name
        rdf += 'ex:' + str(name).replace(' ', '')  + ' a wd:Q177220; \n' 
        rdf += '\t ex:name "' +str(name)+ '"@en;\n'
        #Korean Name
        if not dataset['Korean Name'].isnull()[index]:
          rdf += '\t ex:name "' +dataset['Korean Name'][index]+ '"@kr; \n'
        #Stage Name
        if not dataset['Stage Name'].isnull()[index]:
          rdf += '\t ex:stageName "' +dataset['Stage Name'][index]+ '"@en; \n'
        #Korean Stage Name
        if not dataset['K. Stage Name'].isnull()[index]:
          rdf += '\t ex:koreanStageName "' +dataset['K. Stage Name'][index]+ '"@kr; \n'
        #Date of Birth
        if not dataset['Date of Birth'].isnull()[index]:
          rdf += '\t ex:dateOfBirth "' + str(dataset['Date of Birth'][index])+ '"^^xsd:date; \n' 
        #Country
        if not dataset['Country'].isnull()[index]:
          rdf += '\t ex:country ex:' + replacer(str(dataset['Country'][index])) + '; \n'
        #BirthPlace
        if not dataset['Birthplace'].isnull()[index]:
          rdf += '\t ex:birthPlace ex:' + replacer(str(dataset['Birthplace'][index])) + '; \n'
        #Groups
        if not dataset['Group'].isnull()[index]:
            rdf += element_in_group_idols(dataset['Group'][index], bands_ids_dataset, band_ids_dataset2,  "group")
        if not dataset['Other Group'].isnull()[index]:
            rdf += element_in_group_idols(dataset['Other Group'][index], bands_ids_dataset, band_ids_dataset2, "otherGroup")
        #Gender
        if not dataset['Gender'].isnull()[index]:
            if dataset['Gender'][index] == 'M':
              rdf += '\t ex:gender ex:male. \n' 
            else:
              rdf += '\t ex:gender ex:female. \n' 
        rdf = point_end(rdf)
        index += 1
    return rdf
````

````python
#Obtiene el código de wikidata para una banda y arma la linea rdf correspondiente
def element_in_group_bands(element, dataset, dataset2, typeID):
    #element es el nombre del grupo que se esta buscando
    # dataset es el csv de las bandas
    # dataset2 esd el dataset que contiene los codigos de las bandas q no se encontraron automaticamente pero si a mano
    # typeID es el codigo de girlband o de boyband segun corresponda
    clear_element = element.replace(" ", "")

    if element in dataset['label'].tolist(): #se busca el nombre en el csv
        id_row = (dataset.loc[dataset['label'] == element])['wd_id'].values[0] #si esta se obtiene el código
        text = 'wd:'+ id_row + ' a wd:' + typeID + '; \n' #se pasa a RDF
    elif clear_element in dataset2['label'].tolist():
        id_row = (dataset2.loc[dataset2['label'] == clear_element])['wd_id'].values[0]
        text = 'wd:'+ id_row + ' a wd:' + typeID + '; \n'
    else: #si no está en ningun dataset entonces no se linkea a wikidata, hay un poquito de hardcode, pero porque 
        #era mucha lata reemplazar toooodo, total era solo un caso
        text = 'ex:'+ replacer(clear_element) + ' a wd:' + typeID + '; \n'
    return text

def groupToRDF(dataset, IDtype, ids_dataset, ids_dataset2):
    rdf = ''
    index = 0 #solver for index problem
    for name in dataset['Name']: 
        #id
        rdf += element_in_group_bands(name, ids_dataset, ids_dataset2, IDtype)
        #GroupName
        rdf += '\t ex:name "' +str(name)+ '"@en;\n'
        #Short name
        if not dataset['Short'].isnull()[index]:
          rdf += '\t ex:shortName "' + str(dataset['Short'][index]) +'"^^xsd:string; \n'
        #Korean name
        if not dataset['Korean Name'].isnull()[index]:
          rdf += '\t ex:name "' +dataset['Korean Name'][index]+ '"@kr; \n'
        #Debut
        if not dataset['Debut'].isnull()[index]:
          rdf += '\t ex:debut "' + str(dataset['Debut'][index]) + '"^^xsd:date; \n'
        #Company
        if not dataset['Company'].isnull()[index]:
          rdf += '\t ex:company ex:' + replacer(dataset['Company'][index]) +'; \n'
        #Members
        if not dataset['Members'].isnull()[index]:
          rdf += '\t ex:numberOfActiveMembers "' + str(dataset['Members'][index]) + '"^^xsd:int; \n' 
        #Number of original members
        if not dataset['Orig. Memb.'].isnull()[index]:
          rdf += '\t ex:numberOfInitialMembers "' + str(dataset['Orig. Memb.'][index]) + '"^^xsd:int; \n'
        #Fanclub name
        if not dataset['Fanclub Name'].isnull()[index]:
          rdf += '\t ex:fanclubName ex:' + replacer(str(dataset['Fanclub Name'][index])) +'; \n'
        #Is the group active?
        if not dataset['Active'].isnull()[index]:
          if dataset['Active'][index] == 'Yes':
            rdf += '\t ex:isActive "true"^^xsd:boolean; \n' 
          else:
            rdf += '\t ex:isActive "false"^^xsd:boolean; \n' 
        rdf = point_end(rdf)
        index += 1           
    return rdf
````
